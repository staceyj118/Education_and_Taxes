```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Indiana_k12.csv"
k12_df=pd.read_csv(csv_file)
```

```python
#k12 name, address, city - to json 
k12info = k12_df[['NAME','ADDRESS','CITY']]
k12info
```

```python
#k12 city zip
k12city = k12_df[['CITY','ZIP']]
k12city
```

```python
#k12 district id, name, enrollment 
k12enrollment = k12_df[['DISTRICT_ID', 'NAME', 'ENROLLMENT']]
k12enrollment
```

```python
#k12 start, end, district id
k12grades = k12_df[['START_GRADE','END_GRADE','DISTRICT_ID']]
k12grades
```

```python
rds_connection_string = "postgres:4902@localhost:5432/education_taxes_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
k12info.to_sql(name='k12_info', con=engine, if_exists="append", index=False)
```

```python
pd.read_sql_query('select * from k12_info', con=engine).head()
```

```python
k12city.to_sql(name='k12city', con=engine, if_exists="append", index=False)
```

```python
pd.read_sql_query('select * from k12city', con=engine).head()
```

```python
k12enrollment.to_sql(name='k12enrollment', con=engine, if_exists="append", index=False)
```

```python
pd.read_sql_query('select * from k12enrollment', con=engine).head()
```

```python
k12grades.to_sql(name='k12grades', con=engine, if_exists="append", index=False)
```

```python
pd.read_sql_query('select * from k12grades', con=engine).head()
```
