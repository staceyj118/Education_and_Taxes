```python
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV into DataFrame

```python
csv_file = "Public_Schools.csv"
k12db=pd.read_csv(csv_file)
k12db.head()
```

# Keep relevant columns

```python
k12db=k12db[["NAME", 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'LEVEL_', 'ENROLLMENT', 'ST_GRADE', 'END_GRADE', 'DISTRICTID']]
k12db
```

# Rename Columns

```python
k12db=k12db.rename(columns={'LEVEL_':'LEVEL', 'ST_GRADE':'START_GRADE', 'DISTRICTID':'DISTRICT_ID'})
k12db
```

# Sort by State

```python
sorted_df=k12db.sort_values(by=['STATE'], ascending=True)
sorted_df
```

# Filter to show only Indiana

```python
indiana=sorted_df.loc[sorted_df["STATE"]=='IN']
indiana
```

# Delete schools with 0 enrollment 

```python
indiana=indiana[indiana.ENROLLMENT !=0].reset_index()
indiana_db=indiana[["NAME", 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'LEVEL', 'ENROLLMENT', 'START_GRADE', 'END_GRADE', 'DISTRICT_ID']]
indiana_db
```

```python
indiana_db.to_csv("Indiana_k12.csv")
```
